# Immunology Background

Immunology is the study of the immune system, a network of organs, cells, and proteins that protects the body from foreign invaders ([Hopkins Medicine](https://www.hopkinsmedicine.org/health/conditions-and-diseases/the-immune-system#:~:text=What%20is%20the%20immune%20system,and%20proteins%20that%20work%20together.)). These foreign invaders can be infections, viruses, or any unknown substance that has the potential to harm the body. Understanding the body’s immune response can help researchers better understand how to treat diseases and develop vaccines ([Immunology](https://www.immunology.org/public-information/what-is-immunology)).

<br>

The immune system acts via two types of mechanisms that have evolved over time: innate immunity and adaptive immunity. You can think of these two mechanisms as the body’s soldiers, with innate immunity being the first line of defense and adaptive immunity being the second line of defense. Innate immunity consists of a nonspecific mechanism in which the body provides the same response regardless of which foreign pathogen is invading. Innate immunity utilizes physical barriers such as the skin or saliva, as well as cells which physically protect the body for the first days of infection. If the infection persists, then the second line of defense- adaptive immunity- takes over. Adaptive immunity uses antibodies and T cells to target and kill the foreign pathogens and any infected cells ([Immunology](https://www.immunology.org/public-information/what-is-immunology)). Antibodies are specialized types of proteins created by immune cells that circulate in the bloodstream. They are specific to a certain foreign pathogen (also known as an “antigen”) and will recognize the antigen to target it for destruction and will remain in the bloodstream after initial infection ([Genome](https://www.genome.gov/genetics-glossary/Antibody)).

<br>

Immunoglobulins are antibodies that your body makes in order to fight off foreign invaders within the immune system. There are five different types of Immunoglobulins (Ig) that the body can produce: Immunoglobulin A (IgA), Immunoglobulin G (IgG), Immunoglobulin M (IgM), Immunoglobulin E (IgE), and Immunoglobulin D (IgD). ([KidsHealth](https://kidshealth.org/en/parents/test-immunoglobulins.html)).  Immunoglobulin G (IgG) are the most common antibodies, making up approximately 75% of the total immunoglobulin in humans ([ScienceDirect](https://www.sciencedirect.com/science/article/pii/B012226765600325X)). These antibodies can be found in bodily fluids, as well as the blood, and help fight off certain infections ([KidsHealth](https://kidshealth.org/en/parents/test-immunoglobulins.html)).  Immunoglobulin M is the first antibody  that is secreted by B cells in response to foreign invaders within the body ([NCBI](https://www.ncbi.nlm.nih.gov/books/NBK555995/)). These antibodies are found mainly in the blood and lymph fluid ([KidsHealth](https://kidshealth.org/en/parents/test-immunoglobulins.html)).

# COVID-19 Background

COVID-19 (a type of coronavirus) is a disease that was discovered in December 2019 caused by the SARS-CoV-2 virus ([Röltgen et al., 2020](https://www.science.org/doi/pdf/10.1126/sciimmunol.abe0240)). This coronavirus spreads through droplets that an individual projects out of their mouth after breathing, talking, sneezing, or coughing. SARS-CoV-2 binds a receptor in the host cell in order to enter the cell membrane and to infect the host. Once the host is infected, the viral RNA is released for replication and translation by the cell, which eventually can lead to exocytosis of new viral particles ([Beyerstedt et al., 2021](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7778857/#:~:text=ACE2%20is%20a%20key%20regulator,diabetes%20mellitus%2C%20and%20cardiovascular%20disease.)).  When an individual contracts this virus, some symptoms they may have are fever,  chills, cough, headache, loss of taste or smell, sore throat, nausea or vomiting, and fatigue. While symptoms vary from person to person, those listed above are the most common ([CDC](https://www.cdc.gov/coronavirus/2019-ncov/prevent-getting-sick/prevention.html)).  There are two main types of testing that can be done to detect COVID-19: Viral testing and Antibody testing. Viral testing is done with a sample from the nose or mouth ([CDC](https://www.cdc.gov/coronavirus/2019-ncov/prevent-getting-sick/prevention.html)). For antibody testing, Immunoglobulin (Ig) antibodies that bind the virus are tested.

After that we will count the records using the `.value_counts()` method. And, finally, we will us the Pandas dataframe built-in plotting function `.plot.pie` to create the pie plot. Each of the four steps will be outlined in the sections below:    

<br>

**Step 1:** Filter the data needed. Here, we only need the `Patient_ID` and `Status` columns from the `ip_titer_df` dataframe. We will use the following line of code to filter these two columns, so that we are only looking at these two columns.

`ip_titer_df[['Patient_ID', 'Status']]`


Run the code below.

We will use the `.groupby()` function to perform the grouping using the following line of code:
```python
status_df = ip_titer_df[['Patient_ID', 'Status']].groupby(by=['Patient_ID', 'Status']).mean().reset_index()
```

Notice that there is a bit of code following `.groupby(by=['Patient_ID', 'Status'])`, let's break down what they are:

*  `.mean()` - anytime we group data, we need to provide an aggregate function in case we need to collapse a bunch of numbers or values into a single value. In this case, we don't have any numbers or values to worry about, but we still need to provide the aggregate function, so we put `.mean()` as a placeholder.

*  `.reset_index()` - when we run `.groupby()`, the column values that we used for grouping will turn into the index for each row, in this case, `Patient_ID` and  `Status`. Running `.reset_index()` is to move the index values back into the dataframe as column values.

Now, run the code below.

````python
# Group by patient ID and status (ICU/Admit)
status_df = (ip_titer_df[['Patient_ID', 'Status']]
             .groupby(by=['Patient_ID', 'Status'])
             .mean().reset_index())

# Preview subset
status_df
````

**Step 3:** Count the number of patients for each unique status in the data. In other words, count the number of records for each unique value in the `Status` column. We can use the `.value_counts()` method applied to the Pandas Series `.status_df['Status']`. We then use `.to_frame()` to turn the Pandas Series into a Pandas dataframe.

Run the code below.

````python
# Count the occurrences of each patient status
status_counts = status_df['Status'].value_counts().to_frame()

# Display counts
status_counts
````

**Step 4:** Finally, we will use the dataframe containing the counts of each status to visualize the distribution in a pie chart. We will use the Pandas dataframe built-in plotting function `.plot.pie` to create the pie plot. There are several parts in this function:
* `autopct="%.1f%%"`: This displays the percentage in one decimal point on the pie chart.
* `colors=palette`: This sets the colors of the pie chart to the color we defined in the first line.

Run the code below to display your pie chart.

````python
# Define the colors for the pie chart
palette = ['lightblue','teal']

# Plot the distribution of the patient status
status_counts.plot.pie(y='Status',
                       autopct="%.1f%%", colors=palette)
````

# **Part 6: Visualizing Data using a Boxplot**

A boxplot (or box-and-whisker plot) is a type of graph that allows a visual summary of data for a variable.  The plot contains a box (orange in the example below), whiskers (blue in the example below), and potentially outliers (asterisk in the example below).  The horizontal line within each box is the median (Whitlock & Schluter, 2020).  To determine the median, you would place all the numbers in chronological order and determine the middle value. Thus, 50% of values are below the median, and 50% are above the median. The region between the lower and upper edges of the box represent the middle 50% of values. Variability refers to how spread out the data is. One measure of variability is the range, which is the difference between the smallest and largest numbers in the dataset. In the box plot, this is represented by the whiskers and does not include outliers ([StatisticsHowTo](https://https://www.statisticshowto.com/variability/)).

## **Using Boxplots to visualize IgG and IgM Antibody responses over time:**

**Concept Explained:** In this exercise, you will be creating boxplots to show the development of anti-SARS-CoV2 IgG and IgM antibody responses in inpatients over time. The X-axis displays the days after symptom onset, and the Y-axis represents the absorbance, which indicates the SARS-CoV2 antibody concentration ([Thiha et al., 2015](https://https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4481904/)).

<br>

**Code Explained:** Using the previously generated Pandas dataframe, the [Seaborn](https://seaborn.pydata.org/) library is used to visualize the data using a boxplot. Run the code cells below to see the plot for each antibody type. Let's break down the code below line by line:


The first section of code creates a boxplot using the `seaborn` library. It assigns the resulting plot to the variable `g`. The boxplot is based on the DataFrame `ip_titer_df` filtered for rows where the column `Ig` has the value `'RBD_IgG_Mean'`.
* `x=` is used to label the x-axis of the plot and represent the 'Days Post Symptoms Onset' column.
* `y=` is used to label the y-axis and represents the 'Absorbance (450nm)' column.
*`kind=` is the parameter where we specify that we want a boxplot.
* `aspect=` is a parameter that adjusts the width-to-height ratio of the plot. In this context, an aspect value of 3 means that the plot's width will be three times larger than its height.
* `color=` is a parameter that sets the color displayed, in this case, light blue.


The next line, `g.set_xticklabels(rotation=90)`, rotates the x-axis labels of the plot by 90 degrees. It modifies the x-axis tick labels of the previously created g plot to be displayed vertically, improving readability when the labels are long or overlapping.


The last bit of code uses `.suptitle()` to set the title for the plot. It assigns the title 'Development of anti-SARS-CoV-2 IgG responses in inpatients over time' to the figure (plot) represented by `g.fig`. The `y=1.05` parameter adjusts the vertical position of the title, placing it slightly above the top of the plot.

````python
# Boxplot: Development of anti-SARS-CoV-2 IgG response across time
g=sns.catplot(
    data=ip_titer_df[ip_titer_df.Ig=='RBD_IgG_Mean'],
    x='Days Post Symptoms Onset',
    y='Absorbance (450nm)',
    kind='box',
    aspect=3,
    color='lightblue')

# Rotate x axis label
g.set_xticklabels(rotation=90)

# Designate title and postition
g.fig.suptitle(
    'Development of anti-SARS-CoV-2 IgG responses in inpatients over time',
    y=1.05)
````

````python
# Boxplot: Development of anti-SARS-CoV-2 IgM response across time
g=sns.catplot(
    data=ip_titer_df[ip_titer_df.Ig=='RBD_IgM_Mean'],
    x='Days Post Symptoms Onset',
    y='Absorbance (450nm)',
    kind='box',
    aspect=3,
    color='coral')

# Rotate x axis label
g.set_xticklabels(rotation=90)

# Designate title and postition
g.fig.suptitle(
    'Development of anti-SARS-CoV-2 IgM responses in inpatients over time',
    y=1.05)
````
